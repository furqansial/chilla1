```python
#body mass index (BMI) is a measure of body fat based on height and weight.
#input weight
#input height
#BMI (Ask yourself your height)
# Weight
# Calculate BMI
# Print BMI (my name is ___, and my BMI is ___)
```

```python
def calc_BMI(weight, height):
    BMI = weight / (height ** 2)
    return BMI
```

```python
name=input("What is your name? ")
weight=float(input("What is your weight? "))
height=float(input("What is your height? "))
BMI=calc_BMI(weight, height)
print('My name is {0}'.format(name), "My BMI is", BMI)
```

```python
#Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#for categorical variables
sns.set_theme(style='ticks',color_codes=True)

titanic=sns.load_dataset('titanic')
sns.catplot(x='sex',y='survived',hue='class',kind='bar',data=titanic)
plt.show()
```

```python
# count plot
sns.set_theme(style='whitegrid', color_codes=True)
titanic=sns.load_dataset('titanic')
p1=sns.countplot(x='sex', data=titanic, hue='class')
p1.set_title('Plot for counting')
plt.show()
```

```python
# Scatter plot
g=sns.FacetGrid(titanic, row='sex', hue='alone')
g=(g.map(plt.scatter, 'age', 'fare').add_legend())
plt.show()
```

```python
p2=sns.countplot(x='who', data=titanic, hue='alone')
p2.set_title('Plot for counting')
plt.show()
```
