### facet wrap and facet grid?

```python
# facetgrid seaborn ka aik function hai.. jo k humein canvas per humein grid draw kr k deta hai.
# by default facet frid data ko draw nhi krta... graph per data show krne k liye data ko map kiya jata hai
# and Voilla! your graph with data on it .. is ready to show

tips=sns.load_dataset('tips')
g=sns.FacetGrid(tips, col='time')

# drawing histogram
g.map(plt.hist, 'tip')
plt.show()
```

```python
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(plt.scatter,"tip", "total_bill")
plt.show()
```

### Facet wrap

```python
# it is an argument which is used in FacetGrid. iss argument ki wajah se hum multiple graphs ko hum virtually 
# horizontally ya veritcally show kr skte hain.. or even hum category ki base per hum graph ko show grid k tor per
# show kr skte hain

g=sns.FacetGrid(tips, col='sex', col_wrap=1)

# drawing histogram
g.map(plt.hist, 'tip')
plt.show()
```
