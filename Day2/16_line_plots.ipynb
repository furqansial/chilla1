```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# import dataset
phool=sns.load_dataset('iris')
phool
```

```python
# draw a line plot
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
```

# Adding titles

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.title('Phoolon ka plot')
plt.show()
```

## Adding limits

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.title('Phoolon ka plot')
plt.xlim(2)
plt.ylim(1)
plt.show()
```

# Set styles
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
sns.set_style(style=None, rc=None)
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.title('Phoolon ka plot')
plt.xlim(2)
plt.ylim(1)
sns.set_style('white')
plt.show()
```

## size of figure

```python
#change figure
plt.figure(figsize=(4,2))
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.title('Phoolon ka plot')
plt.show()
```
